```python
cor1 = 'azul'
cor2 = 'rosa'
```

```python
print('O céu é {} e o Ipê é {}'.format(cor1, cor2))
```

```python
print('O céu é {0} e o Ipê é {1}%'.format(cor1, cor2))
```

```python
conta = 17/3

conta
```

```python
print('O resultado da conta é: {0}'.format(conta))
```

```python
print('O resultado da conta é: {:.2f}'.format(conta))
```

```python
print('O resultado da conta é: {0:0.2f}'.format(conta))
```

```python
print('O resultado da conta é: {0:0.4f}%'.format(conta))
```

```python
num_true = 268
num_false = 500

print('Número de casos verdadeiros: {0} ({1:.2f}%)'.format(num_true, (num_true / (num_true + num_false)) * 100))
print('Número de casos falsos: {0} ({1:.2f}%)'.format(num_false, (num_false / (num_true + num_false)) * 100))
```
