## Prevendo se uma pessoa pode ter uma doença cardíaca ou não

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importando o DataFrame

df = pd.read_csv('heart.csv')
```

```python
df.head()
```

## Tratamento e transformação dos Dados

```python
# excluindo colunas indesejáveis

df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
# verificando o tipo dos dados

df.dtypes
```

```python
# Neste caso a Variável "ChestPain" e a variável "Thal" são multiclasse

# vou aplicar uma codificação : LabelEncoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# copiar o data Frame

df_enc = df
df_enc.head(2)
```

```python
# aplicando a codificação nas variáveis desejadas

df_enc['ChestPain'] = le.fit_transform(df_enc['ChestPain'])
```

```python
df_enc.head()
```

```python
# aplicando a codificação na variável "Thal"

df_enc['Thal'] = le.fit_transform(df_enc['Thal'])
```

```python
# Para codificar, a variável Thal deve ser do tipo string ou número, mas ela está em object

# Convertendo o formato da variável para string

df_enc['Thal'] = df_enc['Thal'].astype(str)
```

```python
# Aplicando a codificação da variável novamente

df_enc['Thal'] = le.fit_transform(df_enc['Thal'])
```

```python
df_enc.head()
```

```python
df_enc.dtypes
```

```python
# As variáveis ChestPain e Thal estão em formato diferente das outras inteiras

# É preciso transformar para int64

from numpy import int64
```

```python
df_enc['ChestPain'] = df_enc['ChestPain'].astype(int64)
df_enc['Thal'] = df_enc['Thal'].astype(int64)
```

```python
df_enc.dtypes
```

```python
# Convertendo a variável alvo - (Target) para 0 ou 1 (Classificação binária)

# Onde for Sim: 1, Onde for não: 0

map_apply = {'Yes' : 1, 'No' : 0}

df_enc['AHD'] = df_enc['AHD'].map(map_apply)
```

```python
df_enc['AHD'].head()
```

```python
# Verificando se os formatos estão OK

df_enc.dtypes
```

```python
# Uma Vez que os dados estão formatados, vou verificar se existe nulos (Missing Values)

df_enc.isnull().values.any()
```

```python
# Existem Missing Values!

# Verificando em qual colunas eles estão

for column in df_enc.columns:
    if df_enc[column].isnull().values.any():
        print('Missing Values em: {0}'.format(str(column)))
```

```python
# Existem Missing values somente em Ca

# Verificando o tipo dos dados da coluna
df_enc['Ca'].head()
```

```python
# Tratando Missing Values e transformando para 0.0

df_enc['Ca'] = df_enc['Ca'].fillna(0.0)
```

```python
# Verificando o resultado: Se existem nulos no dataFrame

df_enc.isnull().values.any()
```

```python
df_enc.dtypes
```

## Dividindo os dados em Teste e Treinamento

```python
# Definindo as variáveis de entrada - Features e a variável Target

X = df_enc.drop('AHD', axis=1).values

y = df_enc['AHD'].values
```

```python
# Dividindo os dados utilizando cross Validation

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

# dividindo os dados em folds

kfold = KFold(n_splits=10, shuffle=True, random_state=7)
```

## Escolha e treinamento do modelo

```python
# Como trata-se de um problema de classificação
# Vou utilizar LogisticRegression, SVM e Kneighbors(KNN)
# Esses modelos precisam de normalização ou de padronização

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

lor_model = LogisticRegression()
svm_model = SVC()
knn_model = KNeighborsClassifier()
```

## Testando os modelos com padronização

```python
# importando StandardScaler e padronizando os dados

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scale = scaler.fit(X)
X = scale.transform(X)
```

```python
# Resultado Logistic Regression

lor_result = cross_val_score(lor_model, X, y, cv= kfold)

print('Acurácia: {0:.2f}%'.format(lor_result.mean() * 100))
```

```python
# Resultado Suport Vector Machine

svm_result = cross_val_score(svm_model, X, y, cv= kfold)

print('Acurácia: {0:.2f}%'.format(svm_result.mean() * 100))
```

```python
# Resultado KNN

knn_result = cross_val_score(knn_model, X, y, cv=kfold)

print('Acurácia: {0:.2f}%'.format(knn_result.mean() * 100))
```

## Testando os modelos com Normalização

```python
# importando MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scale = scaler.fit(X)
X = scale.transform(X)
```

```python
# Resultado Logistic Regression

lor_result = cross_val_score(lor_model, X, y, cv= kfold)

print('Acurácia: {0:.2f}%'.format(lor_result.mean() * 100))
```

```python
# Resultado Suport Vector Machine

svm_result = cross_val_score(svm_model, X, y, cv= kfold)

print('Acurácia: {0:.2f}%'.format(svm_result.mean() * 100))
```

```python
# Resultado KNN

knn_result = cross_val_score(knn_model, X, y, cv=kfold)

print('Acurácia: {0:.2f}%'.format(knn_result.mean() * 100))
```

### Neste caso a Normalização gerou um resultado melhor em relação á Padronização
### E o modelo final escolhido foi o LogisticRegression

```python
# Savando do modelo

import joblib

joblib.dump(lor_model, 'LogisticRegression.sav')
```

```python
# Carregando o modelo

loaded_model = joblib.load('LogisticRegression.sav')
```

```python
# Fazendo previsões como o modelo carregado

loaded_model.fit(X, y.ravel())
```
